```python
# https://colab.research.google.com/ 사용


!pip install selenium


!apt-get update -qq
!apt install -y -qq chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import sys
sys.path.insert(0, '/usr/lib/chromium-browser/chromedriver')

from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# 크롬 드라이버 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저 창을 띄우지 않고 실행하려면 이 옵션 사용

# 구글 Colab에서는 브라우저 실행 불가능하므로 아래 옵션 사용
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 크롬 드라이버 실행
driver = webdriver.Chrome(options=options)

# 웹 사이트 접속
url = 'https://kr.investing.com/news/economy/'
driver.get(url)

# 페이지 소스 가져오기
page_source = driver.page_source

# BeautifulSoup으로 파싱
soup = BeautifulSoup(page_source, 'html.parser')

# 경제 메인 페이지의 상위 N개 기사 추출
article_page = soup.find('div', class_='largeTitle')
paragraphs = article_page.find_all('article', "js-article-item")[:17]

top5_links = []
art = []

for link in paragraphs:
  article = link.find('a')
  title = link.find('div', 'textDiv').find('a')['title']
  url = article['href'][13:]
  # print(title)
  # print(url)
  top5_links.append(url)

  ######################
  # 크롬 드라이버 설정
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')  # 브라우저 창을 띄우지 않고 실행하려면 이 옵션 사용

  # 구글 Colab에서는 브라우저 실행 불가능하므로 아래 옵션 사용
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')

  # 크롬 드라이버 실행
  driver = webdriver.Chrome(options=options)

  # 웹 사이트 접속
  url = 'https://kr.investing.com/news/economy'+url
  driver.get(url)

  # 페이지 소스 가져오기
  page_source = driver.page_source

  # 크롬 드라이버 종료
  driver.quit()

  # BeautifulSoup으로 파싱
  soup = BeautifulSoup(page_source, 'html.parser')

  # articlePage 클래스 안에 있는 모든 <p> 태그 추출
  # article_page = soup.find('div', class_='articlePage')
  # paragraphs = article_page.find_all('p')

  #0813 수정 : 중간 광고 Article page not found 처리
  article_page = soup.find('div', class_='articlePage')
  if article_page:
    paragraphs = article_page.find_all('p')
    # rest of your code
  else:
    print("Article page not found")

  sentence = ''
  # 각 <p> 태그의 텍스트를 출력
  for paragraph in paragraphs:
    sentence += paragraph.text

  art.append(sentence)

print(top5_links)
for i in art:
  print(i)
  print()
```
